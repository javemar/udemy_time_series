### Missing Data

Options depends on the  forecasting method:
    
   1. Keep the missing data
   2. Drop the missing data
   3. Fill the missing data

```python
df = pd.DataFrame({"A":[1, 2, np.nan], "B":[5, np.nan, np.nan], "C":[1, 2, 3]})
```

```python
df
```

```python
df.dropna()#drop any row that has missing values
```

```python
df.dropna(axis=1) ## drop any columns that has missing values
```

```python
df.dropna(thresh=2) # set up the minimum number of missing values that are needed to remove a row
```

```python
df.fillna(0) #you can fill values of cells that are null values
```

```python
df.mean()
```

```python
df.fillna(df.mean()) #Filling values with the mean value of each column
```

```python
df["A"].fillna(4) #you can fill values with a individual column
```

### Grouping

Grouping consists in three steps:

* Split
* Apply
* Combine

```python
data   = pd.read_clipboard()
```

```python

data.head()
```

```python
data.shape
```

```python
data.groupby("MUNICIPIO")[["DEPARTAMENTO", "ELEV"]].mean().head() # only take in account columns that have a number data type
```

```python
data.groupby("MUNICIPIO")[["DEPARTAMENTO", "ELEV"]].mean().describe()
```

```python
data.groupby("MUNICIPIO").mean().describe().T
```
